```python
"""
날짜 : 2022/04/28
이름 : 박종우
내용 : Pandas 연습문제3 데이터 그룹화
"""
import pandas as pd
```

```python
# 1-1 데이털르 로드하고 상위 3개를 출력하시오.
data_url = "https://raw.githubusercontent.com/chhak/dataset/master/AB_NYC_2019.csv"
df = pd.read_csv(data_url)
df
```

```python
# 1-2 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하시오.
df.groupby('host_name').size().head()
```

```python
# 1-3 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만드시오.
# 빈도수 컬럼은 counts로 명명하시오.
df.groupby('host_name').size().to_frame().rename(columns={0:'counts'}).sort_values('counts', ascending=False)
```

```python
# 1-4 neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하시오.
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
```

```python
# 1-5 neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group의 최댓값을 출력하시오.
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()\
.groupby(['neighbourhood_group'], as_index=False).max()
```

```python
# 1-6 neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하시오.
df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean','var','max','min'])
```

```python
# 1-7 neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하시오.
df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean','var','max','min'])
```

```python
# 1-8 neighbourhood 값과 neighbourhood_group 값에 따른 price의 평균을 구하시오.
df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean()
```

```python
# 1-9 neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하시오.
df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean().unstack()
```

```python
# 1-10 neighbourhood 값과 neighbourhood_group 값에 따른 price의 편균을 계층적 indexing 없이 구하고
# nan 값은 -999 값으로 채워라
df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean().unstack().fillna(-999)
```

```python
# 1-11 데이터 중 neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 
# price값의 평균, 분산, 최대, 최소값을 구하시오.(상위 5개만 출력)
df[df.neighbourhood_group=='Queens'].groupby(['neighbourhood']).price.agg(['mean','var','max','min']).head()
```

```python
# 1-12 데이터 중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 
# 기준으로 각 값의 비율을 구하시오.
df2 = df[['neighbourhood_group', 'room_type']].groupby(['neighbourhood_group', 'room_type']).size().unstack()
df2.loc[:,:] = (df2.values / df2.sum(axis=1).values.reshape(-1,1))
df2
```
