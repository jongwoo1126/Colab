```python
"""
날짜 : 2022/04/26
이름 : 박종우
내용 : 파이썬 Matplotlib 실습하기

Matplotlib
 - 다양한 시각화 그래프를 지원하는 파이썬 시각화 라이브러리
 - 데이터 탐색 또는 분석결과를 시각화하기 위해 사용하는 파이썬 라이브러리
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
############################################################################
# 라인차트
plt.title('line')
plt.plot([1, 3, 4, 9, 10, 17])
plt.show()
```

```python
# x축
plt.title('line2')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 9, 10, 17])
plt.show()
```

```python
# 스타일
plt.title('line2')
plt.plot([10, 20, 30, 40, 50, 60], [1, 3, 4, 9, 10, 17], 'rs:')
plt.grid(True)
plt.show()
```

```python
# 멀티 라인차트
values = np.arange(0, 5, 0.2)
print(values)

plt.title('line4')
plt.plot(values, values, 'r--')
plt.plot(values, values**2, 'g--')
plt.plot(values, values**3, 'b--')
plt.show()
```

```python
# 라인차트 연습문제
flights = sns.load_dataset('flights')
flights

x = flights.drop_duplicates('year')['year']
y = flights.groupby(by='year')['passengers'].agg('sum')

print('x : ', x)
print('y : ', y)

plt.title('flights')
plt.plot(x,y)
plt.show()
```

```python
############################################################################
# 막대차트  
plt.title('bar')
plt.bar(['apple', 'banana', 'orange', 'grape'], [5, 3, 7, 9])
plt.show()
```

```python
# 막대차트 연습문제
titanic = sns.load_dataset('titanic')
titanic

df_age = titanic[['age']]
df_age.loc[(df_age.age >= 0) & (df_age.age < 10), 'gen'] = 0
df_age.loc[(df_age.age >= 10) & (df_age.age < 20), 'gen'] = 10
df_age.loc[(df_age.age >= 20) & (df_age.age < 30), 'gen'] = 20
df_age.loc[(df_age.age >= 30) & (df_age.age < 40), 'gen'] = 30
df_age.loc[(df_age.age >= 40) & (df_age.age < 50), 'gen'] = 40
df_age.loc[(df_age.age >= 50) & (df_age.age < 60), 'gen'] = 50
df_age.loc[(df_age.age >= 60) & (df_age.age < 70), 'gen'] = 60
df_age.loc[(df_age.age >= 70) & (df_age.age < 80), 'gen'] = 70
df_age.loc[(df_age.age >= 80) & (df_age.age < 90), 'gen'] = 80
df_age.loc[(df_age.age >= 90) & (df_age.age < 100), 'gen'] = 90
df_age

df_count = df_age.groupby(by='gen').count()
df_count

plt.title('titanic')
plt.bar(df_count['gen'], df_count['age'])
plt.show()
```

```python
############################################################################
# 산점도
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

plt.title('scatter')
plt.scatter(x, y)
plt.show()
```

```python
# 산점도 연습문제
tips = sns.load_dataset('tips')
tips

x = tips['total_bill']
y = tips['tip']

plt.title('Tips')
plt.scatter(x, y)
plt.xlabel('Total_Bill')
plt.ylabel('Tip')
plt.show()
```

```python
############################################################################
# 박스상자
data1 = np.random.randint(100, size=5)
data2 = np.random.randint(100, size=5)

print('data1 :', data1)
print('data2 :', data2)

plt.title('boxplot')
plt.boxplot([data1, data2])
plt.xticks([1, 2], ['data1', 'data2'])
plt.show()
```

```python
# 박스상자 연습문제
iris = sns.load_dataset('iris')
iris

dataset1 = iris[iris['species'] == 'setosa']['sepal_length']
dataset2 = iris[iris['species'] == 'versicolor']['sepal_length']
dataset3 = iris[iris['species'] == 'virginica']['sepal_length']

plt.title('iris boxplot')
plt.boxplot([dataset1, dataset2, dataset3], labels=['setosa', 'versicolor', 'virginica'])
plt.show()
```

```python
# 3차원 산점도
import plotly.express as px

fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width',z='petal_width', color='species')
fig.show()
```
